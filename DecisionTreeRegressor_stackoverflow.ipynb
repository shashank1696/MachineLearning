```python
data = data.drop(['    '],axis = 1)
```

```python
data['CompanySizeNumber'] = data['CompanySizeNumber'].fillna('new')
#filling NAN values with the median
```

```python
data1['Salary'] = data1['Salary'].astype(np.int64)
#data['CompanySizeNumber'] = data['CompanySizeNumber'].astype(np.int64)
```

```python
import sklearn.model_selection as ms
import sklearn.tree as tree
```

```python
data1.dtypes
```

```python
x_train,x_test,y_train,y_test=ms.train_test_split(x,y,test_size=0.2,random_state=200)
```

```python
clf=ensemble.RandomForestRegressor(n_estimators=300,oob_score=True,n_jobs=-1) #n_jobs -1 takes all the cpu cores to excecute this
```

```python
clf.fit(x_train,y_train)
```

```python
predictions = clf.predict(x_test)
```

```python
pd.Series(clf.feature_importances_,index=x_train.columns).sort_values(ascending=False)
```

```python
clf.oob_score_ ## oob accuracy score
```

```python
for i in [300,400,500]:
    clf=ensemble.RandomForestRegressor(n_estimators=i,oob_score=True,n_jobs=-1)
    clf.fit(x_train,y_train)
    print('oob score for {} trees is {}'.format(i,clf.oob_score_))
```

```python
errors = abs(predictions - y_test)
```

```python
errors
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
mape = np.mean(100 * (errors / y_test))
accuracy = 100 - mape
print('Accuracy:', round(accuracy, 2), '%.')
```
